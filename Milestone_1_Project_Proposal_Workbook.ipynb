```python
athlete_events_clean = athlete_events.copy()
noc_regions_clean = noc_regions.copy()

athlete_events_clean["Medal"] = athlete_events_clean["Medal"].fillna("No Medal")
noc_regions_clean["region"] = noc_regions_clean["region"].fillna("Unknown")
noc_regions_clean["notes"] = noc_regions_clean["notes"].fillna("")
```

```python
import sqlite3

conn = sqlite3.connect("olympics_capstone.db")

athlete_events_clean.to_sql("athlete_events", conn, if_exists="replace", index=False)
noc_regions_clean.to_sql("noc_regions", conn, if_exists="replace", index=False)
```

```python
query = """
SELECT *
FROM athlete_events
LIMIT 10;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT 
    ae.Name,
    ae.Sex,
    ae.Age,
    ae.Team,
    ae.NOC,
    nr.region,
    ae.Year,
    ae.Season,
    ae.Sport,
    ae.Event,
    ae.Medal
FROM athlete_events ae
LEFT JOIN noc_regions nr
    ON ae.NOC = nr.NOC
LIMIT 10;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT 'athlete_events' AS table_name, COUNT(*) AS row_count
FROM athlete_events

UNION ALL

SELECT 'noc_regions' AS table_name, COUNT(*) AS row_count
FROM noc_regions;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT 
    MIN(Year) AS first_year,
    MAX(Year) AS last_year,
    COUNT(DISTINCT Year) AS number_of_years
FROM athlete_events;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT 
    COUNT(DISTINCT ID) AS unique_athletes,
    COUNT(DISTINCT Sport) AS unique_sports,
    COUNT(DISTINCT Event) AS unique_events,
    COUNT(DISTINCT NOC) AS unique_nocs,
    COUNT(DISTINCT Games) AS unique_games
FROM athlete_events;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT 
    Medal,
    COUNT(*) AS count
FROM athlete_events
GROUP BY Medal
ORDER BY count DESC;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT 
    Year,
    Sex,
    COUNT(DISTINCT ID) AS athlete_count
FROM athlete_events
GROUP BY Year, Sex
ORDER BY Year, Sex;
"""

pd.read_sql_query(query, conn)
```

```python
query = """
SELECT 
    nr.region,
    COUNT(*) AS total_medals
FROM athlete_events ae
LEFT JOIN noc_regions nr
    ON ae.NOC = nr.NOC
WHERE ae.Medal != 'No Medal'
GROUP BY nr.region
ORDER BY total_medals DESC
LIMIT 10;
"""

pd.read_sql_query(query, conn)
```
